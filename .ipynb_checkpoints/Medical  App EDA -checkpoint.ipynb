##### Univariate Analysis

```python
# univariate histograms
def histograms(data):
    print("Histograms of Variables")
    data.hist()
    plt.tight_layout()
    plt.show()
```

```python
histograms(data)
```

```python
def boxplot_histplot(data, feature, figsize=(12,7), bins=None):
    print("Boxplots and Hisplots of variable ", feature)
    fig, (ax_box, ax_hist)=plt.subplots(
    nrows=2,
    sharex=True,
    gridspec_kw={"height_ratios":(0.25, 0.75)},
    figsize=figsize)
    
    sns.boxplot(data=data, x=feature, showmeans=True, ax=ax_box,color='pink')
    sns.histplot(data=data, x=feature, bins=bins, ax=ax_hist, pallete='winter') if bins else sns.histplot(data=data,\
                                                                    x=feature, ax=ax_hist)
    
    ax_hist.axvline(data[feature].mean(), color='green', linestyle='--')
    ax_hist.axvline(data[feature].median(), color='black', linestyle='-')
    plt.show()
    
    
```

```python
for col in data.select_dtypes(exclude='O').columns:
    boxplot_histplot(data=data, feature=col)
```

**Observations**
1. Based on the boxplots and histplots skin thickness, insulin and diabetes pedigree function have lots of outliers and need further treatment

```python
def countplot(data, feature):
    print("Countplot for Outcome")
    plt.figure(figsize=(12,7))
    ax=sns.countplot(data=data, x=feature, color='green')
    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate("{:.2g}%".format(100.*y/len(data)), (x.mean(), y), ha='center', va='bottom')
    plt.show()
```

```python
countplot(data=data, feature='Outcome')
```

**Observation**
The countplot shows that the data is highly imbalanced and needs to be balanced before modelling

##### Bivariate Analysis

```python
def cat_num(data, feature1, feature2):
    print("Bivariate Barcharts between {0} and {1}".format(feature1, feature2))
    data.groupby(feature1)[feature2].mean().plot(kind='bar')
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.show()
```

```python
for col in data.select_dtypes(exclude='O').columns:
    cat_num(data=data, feature1='Outcome', feature2=col)
```

**Observations**
Women with more avergae number of pregnancies, Glucose, Insulin, BMI and Age are more likely to have diabetes

```python
def heatmap(data):
    print("Correlation Heatmap of the data")
    plt.figure(figsize=(12,7))
    sns.heatmap(data.corr(), annot=True, cmap='Spectral', vmin=-1, vmax=+1)
```

```python
heatmap(df)
```

**Observations**
Glocose and BMI are strong predictors of diabetes

### Deploy in Streamlit as an APP

```python
%%writefile app.py
import streamlit as st
st.title("The EDA Page")
# import the libraries 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
st.set_option('deprecation.showPyplotGlobalUse', False)
def main():
    
    def data_cleaning(file):
        data=pd.read_csv(file)
        # drop the unnamed column
        data.drop('Unnamed: 0', axis=1, inplace=True)

        # replace 0s with column median
        zerofill=lambda x:x.replace(0, x.median())
        cols=data.columns[1:6]
        data[cols]=data[cols].apply(zerofill, axis=0)

        # encode the Outcome column
        df=data.copy()
        d={"Yes":1, "No":0}
        df['Outcome']=df["Outcome"].map(d)
        return df
    
    # univariate histograms
    def histograms(data):
        st.subheader("Histograms of Variables")
        data.hist()
        plt.tight_layout()
        st.pyplot()
        
    def countplot(data, feature):
        st.subheader("Countplot for Outcome")
        plt.figure(figsize=(12,7))
        ax=sns.countplot(data=data, x=feature, color='green')
        for p in ax.patches:
            x=p.get_bbox().get_points()[:,0]
            y=p.get_bbox().get_points()[1,1]
            ax.annotate("{:.2g}%".format(100.*y/len(data)), (x.mean(), y), ha='center', va='bottom')
        st.pyplot()
    
    #correlations 
    def heatmap(data):
        st.subheader("Correlation Heatmap of the data")
        plt.figure(figsize=(12,7))
        sns.heatmap(data.corr(), annot=True, cmap='Spectral', vmin=-1, vmax=+1)
        st.pyplot()

    data=data_cleaning(file='data.csv')
    plot=st.sidebar.selectbox("Choose the Plot", ('Histogram', 'Countplot', 'Heatmap'))
    if st.sidebar.button('PLOT'):
        if plot=='Histogram':
            histograms(data)
        if plot=='Countplot':
            countplot(data, feature="Outcome")
        if plot=='Heatmap':
            heatmap(data)
            
if __name__=='__main__':
    main()
            
    
```
